```python
pri=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for a in pri:
    price2.append(a.text)
print(len(price2))
```

```python
ex=[]
disp=driver.find_elements(By.XPATH,'//div[@class="fMghEO"]')
for b in disp:
    b=b.text.split("\n")
    #print(b[2])
    bo=b[0].split("|")
    print(bo)
    ram2.append(bo[0])
    storage2.append(bo[1])
    dispsize2.append(b[1])
    b2=b[2].split("|")
    pcamera2.append(b2[0])
    scamera2.append(b2[1])
    batcap2.append(b[3])
    print(bo[1])
```

```python
df4=pd.DataFrame({"brands":brands2,"prices":price2,"Colour":colours2,"URL":urls2,"RAM":ram2,"ROM":storage2,"Front Camera":pcamera2,"Rear Camera":scamera2,"Batary Cap":batcap2,"Display size":dispsize2})
df4
```
