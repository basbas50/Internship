```python

URL3 = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
```

```python
page3 = requests.get(URL3)
```

```python
content3 = BeautifulSoup(page3.content)
content3
```

```python
first =content3.find('tr',class_="rankings-block__banner").text.split()
table = [first]
#india = content3.find('tr',class_="rankings-block__banner")
for i in content3.find_all('tr',class_='table-body'):
    table.append(i.text.strip().split())
table = table[0:10]
table
```

```python
teams = []
for i in table:
    if len(i[2])>3:
        teams.append(i[1]+" "+i[2])
    else:
         teams.append(i[1])
print(teams)
```

```python
matches = []
for i in table:
    if len(i[2])>3:
        matches.append(i[4]) 
    else:
        matches.append(i[3])
print(matches)  
```

```python
points = []
for i in table:
    if len(i[2])>3:
        points.append(i[5]) 
    else:
        points.append(i[4])
print(points)  
```

```python
ratings = []
for i in table:
    if len(i[2])>3:
        ratings.append(i[6]) 
    else:
        ratings.append(i[5])
print(ratings)  
```

```python
df3 = pd.DataFrame({"Teams":teams,"Matches":matches,"Points":points,"Ratings":ratings})
df3
```
